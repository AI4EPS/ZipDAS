```python
from pathlib import Path
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
quality = 200
raw_jpeg = Path("comp_jpeg/data/Ridgecrest_ODH3-2021-06-15 183838Z.h5")
comp_jpeg = Path(f"comp_jpeg/data_compressed_{quality}/Ridgecrest_ODH3-2021-06-15 183838Z.h5")
```

```python
comp_dl = Path("../decompressed_100/Ridgecrest_ODH3-2021-06-15 183838Z.h5.tfci.h5")
```

```python
with h5py.File(raw_jpeg, 'r') as fp:
    raw_data_jpeg = fp["Data"][:]
# vmax = np.max(raw_data_jpeg)
# vmin = np.min(raw_data_jpeg)
```

```python
with h5py.File(comp_jpeg, 'r') as fp:
    comp_data_jpeg = fp["Data"][:]
comp_data_jpeg = comp_data_jpeg - np.mean(comp_data_jpeg)
# comp_data_jpeg = comp_data_jpeg/255.0 * (vmax - vmin) + vmin
```

```python
with h5py.File(comp_dl, 'r') as fp:
    comp_data_dl = fp["Data"][:]
```

```python
nx, nt = raw_data_jpeg.shape
nt = 1024 * 5
```

```python
plt.clf()
fig, axes = plt.subplots(
    3,
    1,
    figsize=(18 * nt / nx, 12),
    sharex=True,
    sharey=False,
    squeeze=False,
    gridspec_kw={"wspace": 0, "hspace": 0},
)
axes[0, 0].imshow(raw_data_jpeg[:nx, :nt], vmin=-1.5, vmax=1.5, cmap="seismic")
axes[1, 0].imshow(comp_data_jpeg[:nx, :nt], vmin=-1.5, vmax=1.5, cmap="seismic")
axes[2, 0].imshow(comp_data_jpeg[:nx, :nt]-raw_data_jpeg[:nx, :nt],vmin=-1.5, vmax=1.5, cmap="seismic")
fig.subplots_adjust(wspace=None, hspace=None)
fig.savefig(f"result_jpeg_{quality}.png", dpi=300, bbox_inches="tight")
```
